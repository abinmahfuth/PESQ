```python
#Import libraries
from scipy.io import wavfile
from pesq import pesq
import IPython
```

Note that the scale of PESQ ranges from 1 to 5, with 1 being the worst quality and 5 being the best quality 

```python
#Process files
Fs1, speech = wavfile.read('New_Speech.wav')
Fs2, mixed = wavfile.read('New_Mixed.wav')
speech = speech[:,0]
speech = speech[0:Fs1*10]
mixed_c = mixed[:,0]
mixed_c = mixed_c[0:Fs2*10]
mixed_n = mixed[:,1]
mixed_n = mixed_n[0:Fs2*10]
```

```python
print('Speech')
IPython.display.Audio(speech, rate=Fs1)
```

```python
print('Noisy mix')
IPython.display.Audio(mixed_n, rate=Fs2)
```

```python
print('Clean mix')
IPython.display.Audio(mixed_c, rate=Fs2)
```

```python
# PESQ on noisy mix
FsN = 16000
print(pesq(FsN, speech, mixed_n, 'wb'))
#print(pesq(FsN, speech, mixed, 'nb'))
```

```python
# PESQ on clean mix
FsN = 16000
print(pesq(FsN, speech, mixed_c, 'wb'))
#print(pesq(FsN, speech, mixed, 'nb'))
```

```python
# PESQ on the same file
FsN = 16000
print(pesq(FsN, speech, speech, 'wb'))
#print(pesq(FsN, speech, mixed, 'nb'))
```
